```python
# Import Dependencies
import pandas as pd
```

```python
# Reference the file where the CSV is located
csvfile = '../Resources/crime_incident_data2017.csv'

# Import the data into a Pandas DataFrame
df = pd.read_csv(csvfile)
```

```python
# look for missing values
df.count()
```

```python
# drop null rows
df2 = df.dropna(how='any')
```

```python
# verify counts
df2.count()
```

```python
# Check to see if there are any values with mispelled or similar values in "Offense Type"
df['Offense Type'].unique()
```

```python
# Combine similar offenses
df2 = df2.replace({'Drug/Narcotic Violations': 'Drug Violations', 'Drug Equipment Violations': 'Drug Violations'})
df2['Offense Type'].unique()
```
